4. 논리 연산과 비교 연산
> 자바의 true/false 소문자 키워드 <br>
> 파이썬 True/False 키워드

```python
# = : 할당 연산자 / 대입 연산자 
name = '유재석'
print(name)

# == : 동등 비교 연산자 - 값이 같을 경우 True
# != : 부정 연산자 - 값이 다를 경우 True
# 실행 순서 : 1 == 1 비교 -> print()
print(1 == 1)
print(1 != 1)
```

```python
# 문자열 타입 비교
print("a" == "A")
print("a" == "a")
```

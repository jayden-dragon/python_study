**미션- 팩토리얼 계산하기**
>반복문을 활용하여 팩토리얼 계산해 보기<br>
for문과 while문을 활용하여 각각 만들어 보기<br>
팩토리얼 n!은 1~n까지의 정수를 모두 곱한 것을 의미한다<br>
n! = 1*2*3*...(n-1)*n

>힌트 : range() 함수를 이용

````python
def factorial() :
    
    n = int(input("n의 값은? : "))
    
    result = 1
    
    for i in range(1, n + 1) :
        result *= i
        
    print(n, '! 은', result)
    
factorial()
````

**미션 - 알파벳, 숫자, 스페이스의 처리**

> 문자열을 조사하여서 알파벳 문자의 개수, 숫자의 개수, 스페이스의 개수를 출력하는 프로그램을 작성하라

````python
statement = input("문자열을 입력하시오: ")

alphas = 0    # 초기화
digits = 0    # 초기화
spaces = 0    # 초기화

for c in statement:
    if c.isalpha():    # c가 알파벳이냐? 
        alphas = alphas + 1
        
    if c.isdigit():    # c가 숫자냐?
        digits = digits + 1
    
    if c.isspace():    # c가 공백이냐?
        spaces = spaces + 1
            
print ("알파벳 문자의 개수=", alphas)
print ("숫자 문자의 개수=", digits)
print ("스페이스  문자의 개수=", spaces)
````

새로 오픈한 가게에서 오픈 이벤트를 진행한다.

방문한 손님 순서대로 10의 배수번째마다 마우스를, 100의 배수번째마다 모니터를 증정한다.

이벤트가 항목이 중복될 경우 모니터만 제공한다.

모니터 수 : 5, 마우스 수 : 50 이고 소진된 항목은 더이상 증정하지 않으며 모두 소진 시 이벤트는 종료된다.

초기 방문자 수를 입력받는다.

출력으로는 이벤트에 해당되는 방문자 순서와 제공받는 물품의 항목이 되도록 하라.


````python
def event():
    
    mouse = 50
    monitor = 2

    cus_num = int(input())
    
    for visit in range(1, cus_num+1):
        if mouse == 0 and monitor == 0:
            break
    
        if visit%10 == 0:
            if visit%100 == 0:
                if monitor == 0:
                    if mouse > 0 :
                        mouse -= 1
                        print(visit, "번째 손님 마우스 당첨입니다.")
                    else :
                        break
                else :        
                    monitor -= 1
                    print(visit, "번째 손님 모니터 당첨입니다.")
            else:
                if mouse == 0:
                    continue
                else :
                    mouse -= 1
                    print(visit, "번째 손님 마우스 당첨입니다.")

                    
event()
````

중간고사를 실시해서 커트라인 점수 이상이면 advanced , 이하이면 basic 반으로 배정된다.

커트라인은 70점이다.

초기 학생 수와 그에 따른 성적을 입력받는다.

출력으로 adv, basic 반의 인원수가 되도록 하라.

````python
def midterm():
    
    cutline = 70
    adv_num = 0
    basic_num = 0

    student_num = int(input())
    score = list(map(int, input().split()))
    
    for i in range(0, student_num):
        if score[i] >= cutline:
            adv_num += 1
        else:
            basic_num += 1
    
    print("advance반의 학생 수는", adv_num)
    print("basic반의 학생 수는", basic_num)
    
    
midterm()
````

임의의 숫자의 범위를 정하고 (예: 0~100) 반복문을 사용하여 범위 내의 소수를 출력하는 프로그램을 작성하시오



````python
def prime_num() : 
    
    value = int(input())

    for i in range(1, value + 1) :
        cnt = 0

    for j in range(1, i + 1) :
        if i % j == 0 :
            cnt += 1
    
    if cnt == 2 :
        print(i, end = ' ')

        
prime_num()
````
